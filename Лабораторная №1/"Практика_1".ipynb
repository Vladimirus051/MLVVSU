**Постройте корреляционную матрицу для всех числовых переменных и визуализируйте её с помощью тепловой карты:**

```python
numeric_cols = ['Age', 'Annual Income ($)', 'Spending Score (1-100)', 'Work Experience', 'Family Size']
correlation_matrix_clean = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_clean,
            annot=True,
            cmap='coolwarm',
            center=0,
            square=True,
            fmt='.2f',
            cbar_kws={'shrink': .8})
plt.title('Корреляционная матрица числовых переменных', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

Корреляционная матрица показывает силу линейной связи между пятью числовыми переменными в датасете.
